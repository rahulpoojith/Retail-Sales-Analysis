```python
#Making Not available and unknown as NaN
df=pd.read_csv("/Users/rahulpoojith/Retail Analysis Order/orders.csv",na_values=['Not Available','unknown'])
```

```python
df.head(20)
```

```python
df.columns=[col.lower().replace(' ', '_') for col in df.columns]
```

```python
df.head(20)
```

```python
#Derive new columns discount, sale price and profit
df['discount']=round(df['list_price']*df['discount_percent']*.01,4)
```

```python
df.head(20)
```

```python
# Now detemine the sale price as we have given a discount, profit or less
df['sale_price']=df['list_price']-df['discount']
df['profit']=round(df['sale_price']-df['cost_price'],3)
```

```python
df.head(20)
```

```python
## Convert the order date from object type to datetime type
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d").dt.date
```

```python
df.head(20)
```

```python
#drop cost price and list price as we have derived the sale price and profit
df.drop(['cost_price','list_price'],axis=1,inplace=True)
```
